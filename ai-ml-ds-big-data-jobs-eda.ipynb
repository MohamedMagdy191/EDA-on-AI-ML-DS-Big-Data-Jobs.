## Importing Libraries

```python
import os
import pandas as pd
import numpy as np 
import re

import matplotlib.pyplot as plt 
import seaborn as sns
```

## Data Preprocessing 

```python
for dirname, _, filenames in os.walk('/kaggle/input'):
    for filename in filenames:
        print(os.path.join(dirname, filename))
```

```python
#load the csv file in dataframe

df = pd.read_csv('/kaggle/input/data-science-jobs-comprehensive-dataset/data_science_job.csv',
                 encoding='ISO-8859-1')
```

```python
#renaming the columns 

df.columns = ['company','job_title','location','job_type','experience_level',
             'salary','requirments','facilities']
#df.head()
```

```python
#view the shape of data

print("Number of instances", df.shape[0])
print("Number of features", df.shape[1])
```

```python
#checking the missing values

print(df.isna().sum())
df = df.dropna()
df = df.reset_index(drop=True)
```

```python
#data types of the columns 

df.dtypes
```

```python
#function for keyword searching

def findWholeWord(w):
    return re.compile(r'\b({0})\b'.format(w), flags=re.IGNORECASE).search
```

```python
#renaming the job titles

df['job_title']  = df['job_title'].str.strip()
df['job_title']  = df['job_title'].str.replace('-', ' ')
df['job_title']  = df['job_title'].str.replace('/', ' ')
df['job_title']  = df['job_title'].str.replace('&', ' ')
df['job_title']  = df['job_title'].str.replace(',', ' ')
df['job_title']  = df['job_title'].str.replace(';', ' ')

for i,n in enumerate(df.job_title):
    
    ml1 = findWholeWord('Machine Learning')(str(n))
    ml2 = findWholeWord('Deep Learning')(str(n))
    ml3 = findWholeWord('Vision')(str(n))
    ml4 = findWholeWord('Natural Language Processing')(str(n))
    ml5 = findWholeWord('Text')(str(n))
    ml6 = findWholeWord('Linguist')(str(n))
    ml7 = findWholeWord('ML')(str(n))
    ml8 = findWholeWord('ChatGPT')(str(n))
   
    
    bi1 = findWholeWord('Business Intelligence')(str(n))
    bi2 = findWholeWord('bi')(str(n))
    bi3 = findWholeWord('BI')(str(n))
    bi4 = findWholeWord('PowerBI')(str(n))
    bi5 = findWholeWord('Power BI')(str(n))
    bi6 = findWholeWord('Business Intel')(str(n))
    
    sw1 = findWholeWord('Software Engineer')(str(n))
    sw2 = findWholeWord('SW')(str(n))
    sw3 = findWholeWord('HTML')(str(n))
    sw4 = findWholeWord('Data Platform')(str(n))
    sw5 = findWholeWord('Backend')(str(n))
    
    ai1 = findWholeWord('AI')(str(n))
    ai2 = findWholeWord('Robotics')(str(n))
    ai3 = findWholeWord('Artificial Intelligence')(str(n))
    
    devops1 = findWholeWord('DevOps')(str(n))
    devops2 = findWholeWord('Operations')(str(n))
    devops3 = findWholeWord('DataOps')(str(n))
    devops4 = findWholeWord('MLOps')(str(n))
    
    data_ana1 = findWholeWord('Data Analyst')(str(n))
    data_ana2 = findWholeWord('Data Analytics')(str(n))
    data_ana3 = findWholeWord('Analytics Engineer')(str(n))
    data_ana4 = findWholeWord('Insight Analyst')(str(n))
    data_ana5 = findWholeWord('Visualization')(str(n))
    data_ana6 = findWholeWord('Research Analyst')(str(n))
    
    data_sci1 = findWholeWord('Data Scientist')(str(n))
    data_sci2 = findWholeWord('Data Science')(str(n))
    data_sci3 = findWholeWord('Data science')(str(n))
    data_sci4 = findWholeWord('Research Scientist')(str(n))
    data_sci5 = findWholeWord('Applied Scientist')(str(n))
    
    d_eng1 = findWholeWord('Data Engineer')(str(n))
    d_eng2 = findWholeWord('Data Engineering')(str(n))
    d_eng3 = findWholeWord('Data Pipeline')(str(n))
    d_eng4 = findWholeWord('ETL')(str(n))
    d_eng5 = findWholeWord('Data Management')(str(n))
    d_eng6 = findWholeWord('Data Developer')(str(n))
    d_eng7 = findWholeWord('Databricks')(str(n))
    d_eng8 = findWholeWord('Cloud')(str(n))
    
    big_data1 = findWholeWord('Bigdata')(str(n))
    big_data2 = findWholeWord('Big Data')(str(n))
    

    data_arch = findWholeWord('Data Architect')(str(n))
    
    data_man = findWholeWord('Data Manager')(str(n))
    
    data_str1 = findWholeWord('Data Strategist')(str(n))
    data_str2 = findWholeWord('Data Strategy')(str(n))
    
    data_spec = findWholeWord('Data Specialist')(str(n))
    
    data_pro = findWholeWord('Data Product')(str(n))
    if ml1 or ml2 or ml3 or ml4 or ml5 or ml6 or ml7 or ml8:
        df.loc[i,"job_title"] = "ML Engineer"
    elif bi1 or bi2 or bi3 or bi4 or bi5 or bi6:
        df.loc[i,"job_title"] = "BI Analyst"
    elif sw1 or sw2 or sw3 or sw4 or sw5:
        df.loc[i,"job_title"] = "Software Engineer"
    elif ai1 or ai2 or ai3:
        df.loc[i,"job_title"] = "AI Engineer"
    elif devops1 or devops2 or devops3 or devops4:
        df.loc[i,"job_title"] = "DevOps Engineer"
    elif data_ana1 or data_ana2 or data_ana3 or data_ana4 or data_ana5 or data_ana6:
        df.loc[i,"job_title"] = "Data Analyst" 
    elif data_sci1 or data_sci2 or data_sci3 or data_sci4 or data_sci5:
        df.loc[i,"job_title"] = "Data Scientist"
    elif d_eng1 or d_eng2 or d_eng3 or d_eng4 or d_eng5 or d_eng6 or d_eng7 or d_eng8:
        df.loc[i,"job_title"] = "Data Engineer"
    elif data_arch:
        df.loc[i,"job_title"] = "Data Architect"
    elif data_man:
        df.loc[i,"job_title"] = "Data Manager"
    elif data_str1 or data_str2:
        df.loc[i,"job_title"] = "Data Strategist"
    elif data_spec:
        df.loc[i,"job_title"] = "Data Specialist"      
    elif data_pro:
        df.loc[i,"job_title"] = "Data Product Owner" 
    elif big_data1 or big_data2:
        df.loc[i,"job_title"] = "Big Data Engineer"
    else:
        df.loc[i,"job_title"] = "Other Data Related Job"
    
```
